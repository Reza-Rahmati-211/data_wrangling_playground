### Exercise 14.3 - cumcount and ngroup

Sort by customer/date then add sequence order via `cumcount()+1` and group id via `ngroup` by channel.

```python
import pandas as pd
df = pd.DataFrame({'customer_id':['C1','C1','C2','C2'],'channel':['web','web','store','web'],'order_date':['2025-01-01','2025-01-05','2025-01-02','2025-01-03']})
df['order_date']=pd.to_datetime(df['order_date'])
```

```python
out = df.sort_values(['customer_id','order_date']).copy()
out['order_no'] = out.groupby('customer_id').cumcount() + 1
out['channel_gid'] = out.groupby('channel').ngroup()
out
```

### Exercise 21.3 - pipe Chain

Use `pipe` to fill missing age and add `age_band` via `cut`.

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'age':[24,np.nan,36,42]})
```

```python
def add_band(d):
    out = d.copy()
    out['age_band'] = pd.cut(out['age'], bins=[0,30,40,100], labels=['young','mid','senior'])
    return out

res = df.pipe(lambda d: d.assign(age=d['age'].fillna(d['age'].median()))).pipe(add_band)
res
```

## 26. End-to-end realistic workflow

### Exercise 26.1 - Clean Raw Events

Clean schema/text, parse dtypes, handle missing sessions/spend.

```python
import pandas as pd
import numpy as np
raw = pd.DataFrame({' Customer ID ':["C001","C002","C002","C003"],'Event Date':['2025-03-01','2025/03/02','2025-03-02','bad_date'],'City':[' ny','SF','sf ','LA'],'Plan':['Basic','PRO','pro','Enterprise'],'Sessions':[5,12,12,np.nan],'Spend':[40.0,120.0,120.0,300.0],'Target':[0,1,1,0]})
```

```python
flow = raw.copy()
flow.columns = flow.columns.str.strip().str.lower().str.replace(' ','_', regex=False)
flow['customer_id'] = flow['customer_id'].str.strip().str.upper()
flow['city'] = flow['city'].str.strip().str.upper()
flow['plan'] = flow['plan'].str.strip().str.lower()
flow['event_date'] = pd.to_datetime(flow['event_date'], errors='coerce')
flow['sessions'] = pd.to_numeric(flow['sessions'], errors='coerce').fillna(flow['sessions'].median())
flow['spend'] = pd.to_numeric(flow['spend'], errors='coerce').fillna(flow['spend'].median())
flow = flow.dropna(subset=['event_date'])
flow
```

### Exercise 26.2 - Feature Engineering + Merge Lookup

Add `spend_per_session`, `is_high_session`, city summary features, and merge region lookup.

```python
import pandas as pd
flow = pd.DataFrame({'customer_id':['C001','C002','C004'],'event_date':pd.to_datetime(['2025-03-01','2025-03-02','2025-03-05']),'city':['NEW YORK','SF','LA'],'plan':['basic','pro','basic'],'sessions':[5,12,12],'spend':[40.0,120.0,95.0],'target':[0,1,1]})
region_lookup = pd.DataFrame({'city_std':['NEW YORK','SF','LA'],'region':['East','West','West']})
```

```python
x = flow.copy()
x['spend_per_session'] = x['spend'] / x['sessions']
x['is_high_session'] = (x['sessions'] >= 12).astype(int)
city_stats = x.groupby('city', as_index=False).agg(city_events=('customer_id','count'), city_avg_spend=('spend','mean'))
x = x.merge(city_stats, on='city', how='left')
x = x.merge(region_lookup, left_on='city', right_on='city_std', how='left').drop(columns=['city_std'])
x
```
