## Why NumPy Matters

NumPy is the foundation of scientific computing in Python because it gives us fast, memory-efficient arrays that are far more suitable for numerical work than plain Python lists.

A Python list is flexible, but each element is a separate Python object. NumPy stores values in a compact, homogeneous block of memory, which makes large-scale computation faster and more predictable.

NumPy also enables vectorization, meaning we can apply operations to entire arrays at once instead of looping element by element in pure Python. That leads to cleaner code, fewer bugs, and major performance gains.

It also matters because much of the Python data stack is built on top of it. Libraries such as pandas, matplotlib, scikit-learn, and many machine learning tools rely on NumPy arrays and NumPy-style computation under the hood.

In practice, learning NumPy means learning the core language of data manipulation, numerical analysis, and model preparation in Python.

```python
def quick_sort(l):
    if len(l) <= 1:
        return l
        
    arr = np.array(l)
    v = arr[len(arr)//2]
    
    s = arr[arr < v]
    mid = arr[arr == v]
    b = arr[arr > v]
    return np.concatenate([quick_sort(s),mid,quick_sort(b)])

h = [3,1,7,5,6,4,9,2,8]
sorted =quick_sort(h)
sorted
```

```python
not_sorted = np.random.randint(1,1000,(1,18))[0]
print(not_sorted)
sorted = quick_sort(not_sorted)
print(sorted)
```

## Coordinate Grids and Surface Construction

```python
x = np.linspace(1,10,10)
y = np.linspace(1,10,10)
X, Y = np.meshgrid(x,y)

print(X)
print(Y)
print(np.meshgrid(x,y))
```

```python
planes = np.array([
    [1,2,1,-2],
    [1,-1,-2,2],
    [2,-2,1,2]
])

x = np.linspace(-10,10,50)
y = np.linspace(-10,10,50)
X, Y = np.meshgrid(x,y)

fig = go.Figure()

colors = ["Viridis","Plasma","Cividis"]

for i in range(3):
    a,b,c,d = planes[i]

    Z = d/c - (a/c)*X - (b/c)*Y

    fig.add_surface(
        x=X,
        y=Y,
        z=Z,
        colorscale=colors[i],
        opacity=0.7,
        showscale=False
    )

fig.update_layout(
    title="Three Planes",
    scene=dict(
        xaxis_title="x",
        yaxis_title="y",
        zaxis_title="z",
        aspectmode="cube"
    )
)

fig.show()

```
