## 2. Why NumPy Matters

NumPy matters because it gives Python a fast, consistent, and expressive way to work with numerical data at scale. If plain Python lists are flexible general-purpose containers, NumPy arrays are specialized tools built for mathematical computing.

A Python list can hold many kinds of objects, which is convenient but less efficient for large numerical workloads. A NumPy array stores values in a homogeneous, compact memory layout, which makes computation faster and more predictable. That difference becomes especially important when datasets grow beyond toy examples.

NumPy also introduces vectorization: instead of writing explicit Python loops for every calculation, we can apply operations to entire arrays at once. This usually leads to code that is shorter, easier to read, and significantly faster because much of the heavy lifting happens in optimized low-level implementations.

Another reason NumPy matters is that it sits underneath much of the modern Python data stack. pandas relies on NumPy-style array structures, matplotlib accepts NumPy arrays naturally, scikit-learn expects NumPy-shaped data, and many machine learning workflows begin with NumPy arrays long before a model is trained.

In practice, learning NumPy is not just about learning one library. It is about learning the core mental model for numerical computing in Python: arrays, shapes, vectorized operations, broadcasting, and matrix-oriented thinking.

Key takeaways:

- NumPy arrays are more efficient than Python lists for numerical work.
- Vectorized operations reduce manual looping and improve performance.
- Many data science and machine learning libraries are built around NumPy concepts.
- Strong NumPy fundamentals make the rest of the Python data ecosystem easier to learn.

## 3. Array Creation Fundamentals

This section shows the core array creation patterns. The goal is to understand which constructor to reach for and what kind of structure it produces.

### Creating Arrays From Python Data

The most direct starting point is converting existing Python sequences into NumPy arrays with `np.array()`.

```python
python_list = [1, 2, 3, 4, 5]
python_tuple = (10, 20, 30, 40)

array_from_list = np.array(python_list)
array_from_tuple = np.array(python_tuple)

print("From list:", array_from_list)
print("From tuple:", array_from_tuple)
print("Type:", type(array_from_list))
print("Dtype:", array_from_list.dtype)
```

### 1D, 2D, and 3D Arrays

Nested sequences define the dimensional structure of the resulting array.

```python
vector_1d = np.array([1, 2, 3, 4])
matrix_2d = np.array([[1, 2, 3], [4, 5, 6]])
tensor_3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print("1D array:\n", vector_1d)
print("shape:", vector_1d.shape, "| ndim:", vector_1d.ndim)
print()
print("2D array:\n", matrix_2d)
print("shape:", matrix_2d.shape, "| ndim:", matrix_2d.ndim)
print()
print("3D array:\n", tensor_3d)
print("shape:", tensor_3d.shape, "| ndim:", tensor_3d.ndim)
```

### Generated Numeric Sequences

Use `np.arange()` for step-based ranges and `np.linspace()` when you want a fixed number of evenly spaced values.

```python
print(np.arange(10))
print(np.arange(1,10))
print(np.arange(2,10,2))

print(np.linspace(1,10,20))

# range_values = np.arange(0, 12, 2) # inclusive at start and exclusive at end
# evenly_spaced = np.linspace(0, 1, 6) # inclusive at start and end

# print("np.arange(0, 12, 2):", range_values)
# print("np.linspace(0, 1, 6):", evenly_spaced)
```

### Arrays With Default Values

These constructors are useful when you already know the shape you need and want NumPy to fill it efficiently.

```python
zeros_array = np.zeros((2, 3))
ones_array = np.ones((2, 3))
full_array = np.full((2, 3), 7)
empty_array = np.empty((2, 3))
identity_array = np.eye(4)

print("Zeros:\n", zeros_array)
print()
print("Ones:\n", ones_array)
print()
print("Full of 7s:\n", full_array)
print()
print("Empty (uninitialized values):\n", empty_array)
print()
print("Identity matrix:\n", identity_array)
```

### Dtypes and Conversions

Controlling `dtype` helps with precision, memory usage, and compatibility with later operations.

```python
int_array = np.array([1, 2, 3, 4], dtype=np.int32)
float_array = np.array([1, 2, 3, 4], dtype=np.float64)
converted_to_float = int_array.astype(np.float64)
back_to_python = float_array.tolist()

print("int_array dtype:", int_array.dtype)
print("float_array dtype:", float_array.dtype)
print("converted_to_float dtype:", converted_to_float.dtype)
print("Back to Python list:", back_to_python)
```

Key takeaways:

- `np.array()` bridges Python sequences and NumPy arrays.
- Nested lists naturally create 2D and 3D structures.
- `arange` and `linspace` solve different sequence-generation problems.
- `zeros`, `ones`, `full`, `empty`, and `eye` are core constructors you will use often.
- `dtype` choices and `astype()` conversions are part of everyday NumPy work.

## 4. Array Inspection and Metadata

This section shows how to read an array quickly by inspecting its structure, type, and memory footprint.

```python
sample_vector = np.array([1, 2, 3, 4, 5], dtype=np.int32)
sample_matrix = np.array([[1.5, 2.5, 3.5], [4.5, 5.5, 6.5]])

for name, arr in {"sample_vector": sample_vector, "sample_matrix": sample_matrix}.items():
    print(name)
    print(arr)
    print("shape:", arr.shape)
    print("ndim:", arr.ndim)
    print("size:", arr.size)
    print("dtype:", arr.dtype)
    print("itemsize:", arr.itemsize)
    print("nbytes:", arr.nbytes)
    print("-" * 30)
```

Quick inspection habits make debugging much easier. If a NumPy result looks wrong, `shape`, `dtype`, and `ndim` are usually the first things to check.

## 5. Shapes, Dimensions, and Reshaping

Reshaping is how we reorganize data without changing the underlying values. This is essential before broadcasting, matrix multiplication, and machine learning preprocessing.

```python
base = np.arange(1, 13)
reshaped = base.reshape(3, 4)
flattened_view = reshaped.ravel()
flattened_copy = reshaped.flatten()
transposed = reshaped.T
column_vector = base.reshape(-1, 1)
expanded = np.expand_dims(base[:4], axis=0)
squeezed = np.squeeze(expanded)

print("base shape:", base.shape)
print("reshaped to 3x4:\n", reshaped)
print("transposed shape:", transposed.shape)
print(transposed)
print("column_vector shape:", column_vector.shape)
print("expanded shape:", expanded.shape)
print(expanded)
print("squeezed shape:", squeezed.shape)
print(squeezed)
print("ravel shares data conceptually as a view when possible:", flattened_view)
print("flatten creates a copy:", flattened_copy)
```

## 6. Indexing, Slicing, and Selection

Indexing is where array fluency becomes practical. Good slicing habits let you pull out rows, columns, windows, and submatrices without extra loops.

```python
grid = np.arange(1, 17).reshape(4, 4)
three_d = np.arange(1, 25).reshape(2, 3, 4)

print("grid:\n", grid)
print("first element:", grid[0, 0])
print("last column:", grid[:, -1])
print("middle block:\n", grid[1:-1, 1:-1])
print("every other row:\n", grid[::2])
print("fancy indexing: (0, 2) and (1, 3) :", grid[[0, 2], [1, 3]] )
print()
print("three_d[1, 2, 3]:", three_d[1, 2, 3])
print("three_d[0, :, 1:3]:\n", three_d[0, :, 1:3])
```

```python
original = np.arange(10)
view = original[2:7]
copy = original[2:7].copy()
view[0] = 999
copy[1] = -5

print("original after editing view:", original)
print("view:", view)
print("copy (independent):", copy)
```

## 7. Vectorized Computation and Broadcasting

This is one of NumPy's biggest advantages. We can express whole-array operations directly and let optimized routines handle the heavy work.

```python
x = np.arange(1, 6)
y = np.array([10, 20, 30, 40, 50])
manual = np.array([value * 2 + 1 for value in x])
vectorized = x * 2 + 1

print("manual result:", manual)
print("vectorized result:", vectorized)
print("x + y:", x + y)
print("x ** 2:", x ** 2)
print("dot product:", np.dot(x, y))
```

```python
matrix = np.arange(1, 13).reshape(3, 4)
row_offsets = np.array([100, 200, 300, 400])
column_scale = np.array([[1], [10], [100]])

print("matrix:\n", matrix)
print("matrix + row_offsets:\n", matrix + row_offsets)
print("matrix * column_scale:\n", matrix * column_scale)
print("broadcasted result shape:", (matrix + row_offsets).shape)
```

## 8. Mathematical Functions and Aggregations

NumPy includes both elementwise mathematical functions and summary operations that reduce arrays to useful statistics.

```python
values = np.array([0, 1, 2, 3, 4], dtype=float)
shifted = values + 1

print("shifted:",shifted)
print("sin:", np.round(np.sin(values), 3))
print("cos:", np.round(np.cos(values), 3))
print("exp:", np.round(np.exp(values), 3))
print("log(shifted):", np.round(np.log(shifted), 3))
print("sqrt(shifted):", np.round(np.sqrt(shifted), 3))
print("abs([-3, -1, 2]):", np.abs(np.array([-3, -1, 2])))
```

```python
scores = np.array([
    [80, 90, 85], 
    [70, 88, 92], 
    [95, 91, 89]])

print("scores:\n", scores)
print("sum:", scores.sum())
print("mean by column:", scores.mean(axis=0))
print("median by row:", np.median(scores, axis=1))
print("std:", np.round(scores.std(), 3))
print("var:", np.round(scores.var(), 3))
print("min and argmin:", scores.min(), scores.argmin())
print("max and argmax:", scores.max(), scores.argmax())
print("cumsum row 0:", np.cumsum(scores[0]))
print("cumprod of [1,2,3,4]:", np.cumprod(np.array([1, 2, 3, 4])))
```

## 9. Boolean Masks and Conditional Logic

Boolean arrays make filtering, thresholding, and rule-based transformations concise and expressive.

```python
temperatures = np.array([18, 21, 25, 29, 31, 16, 24])
warm_mask = temperatures >= 24
comfort_mask = (temperatures >= 20) & (temperatures <= 30)

print("temperatures:", temperatures)
print("warm_mask:", warm_mask)
print("warm values:", temperatures[warm_mask])
print("comfortable values:", temperatures[comfort_mask])
print("any above 30?:", np.any(temperatures > 30))
print("all above 15?:", np.all(temperatures > 15))
```

```python
sensor = np.array([0.2, 0.9, 1.7, 2.5, 3.6, 4.1])
labels = np.where(sensor >= 2.0, "high", "low")
clipped = np.where(sensor > 3.0, 3.0, sensor)

print("sensor:", sensor)
print("labels:", labels)
print("clipped to max 3.0:", clipped)
```

## 10. Sorting, Searching, and Set Operations

NumPy includes practical tools for ordering values, locating interesting positions, and comparing collections of values.

```python
raw = np.array([7, 2, 9, 2, 5, 7, 1])

print("raw:", raw)
print("sorted:", np.sort(raw))
print("argsort:", np.argsort(raw))
print("unique:", np.unique(raw))
print("indices where raw > 5:", np.where(raw > 5)[0])
print("nonzero indices:", np.nonzero(raw))
```

```python
group_a = np.array([1, 2, 3, 4, 5])
group_b = np.array([4, 5, 6, 7])

print("intersection:", np.intersect1d(group_a, group_b))
print("difference a-b:", np.setdiff1d(group_a, group_b))
print("union:", np.union1d(group_a, group_b))
print("membership of [2,6,8] in group_a:", np.isin(np.array([2, 6, 8]), group_a))
```

## 11. Iteration, Enumeration, and Practical Patterns

Vectorization is usually best, but iteration still appears when we need row-wise inspection, custom formatting, or small algorithmic routines.

```python
table = np.arange(1, 13).reshape(3, 4)

print("Iterating over rows:")
for row in table:
    print("row sum:", row.sum(), "| row:", row)

print()
print("Iterating with enumerate:")
for idx, row in enumerate(table):
    print(f"row {idx} mean = {row.mean():.2f}")
```

```python
def moving_average(arr, window):
    result = []
    for i in range(len(arr) - window + 1):
        result.append(arr[i:i + window].mean())
    return np.array(result)

series = np.array([3, 5, 8, 10, 12, 15])
print("series:", series)
print("moving average window=3:", moving_average(series, 3))
```

## 13. Linear Algebra Essentials

NumPy provides the array operations that underpin regression, optimization, geometry, and many machine learning models.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
A = np.array([
    [1, 1, 1], 
    [3, 4, 1],
    [1, 1, 1]
    ])
B = np.array([
    [7, 8, 2],
    [9, 10, 2],
    [11, 12, 2]
    ])

print("dot product:", np.dot(a, b))
print("inner product:", np.inner(a, b))
print("matrix multiplication using @:\n", A @ B)
print("matrix multiplication using np.dot:\n", np.dot(A,B))
print("transpose of B:\n", B.T)
```

```python
M = np.array([
    [2.0, 1.0], 
    [1.0, 3.0]
    ])
target = np.array([1.0, 2.0])
solution = np.linalg.solve(M, target) # solve will raise exception if there is no explicit solution.

print("solution to Mx = target:", np.round(solution, 3))
print("determinant:", np.round(np.linalg.det(M), 3))
print("inverse:\n", np.round(np.linalg.inv(M), 3))
print("rank:", np.linalg.matrix_rank(M))
eigenvalues, eigenvectors = np.linalg.eig(M)
print("eigenvalues:", np.round(eigenvalues, 3))
print("eigenvectors:\n", np.round(eigenvectors, 3))
```

## 14. Coordinate Grids, Simulation Inputs, and Geometry

Coordinate grids make it easy to evaluate formulas across whole 2D domains. This is useful in plotting, simulation, optimization, and geometry.

```python
x = np.linspace(-2, 2, 5)
y = np.linspace(-2, 2, 5)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

print("x: ", x ,"\ny: ", y)
print("X:\n", X)
print("Y:\n", Y)
print("X shape:", X.shape)
print("Y shape:", Y.shape)
print("surface Z = X^2 + Y^2:\n", Z)
```

```python
x_line = np.linspace(0, 4, 5)
y_line = np.linspace(0, 4, 5)
X_plane, Y_plane = np.meshgrid(x_line, y_line)
plane = 2 * X_plane + 3 * Y_plane + 5

print("plane values for z = 2x + 3y + 5:\n", plane)
print("center point value:", plane[2, 2])
```

## 15. Performance Mindset and Memory Awareness

A good NumPy workflow is not only correct; it is also conscious of speed, memory, and whether operations create views or copies.

```python
large = np.arange(1000000, dtype=np.float64)
view = large[:5]
copy = large[:5].copy()

print("large dtype:", large.dtype)
print("large nbytes:", large.nbytes)
print("first five via view:", view)
print("copy shares memory?", np.shares_memory(large, copy))
print("view shares memory?", np.shares_memory(large, view))
```

```python
small = np.arange(1, 3000001)
t = time()
python_sum = sum([value * 2 for value in small])
print("python time: ", time()-t)
t = time()
numpy_sum = np.sum(small * 2)
print("numpy time: ", time()-t)

print("python_sum:", python_sum)
print("numpy_sum:", numpy_sum)
print("same result:", python_sum == numpy_sum)
print("Vectorized NumPy usually wins more clearly as data gets larger.")
```

## 16. NumPy for Machine Learning Foundations

Many machine learning workflows start as plain array operations: building feature matrices, transforming columns, and applying matrix math to predictions.

```python
features = np.array([
    [170, 65],
    [180, 80],
    [175, 72],
    [160, 55]
], dtype=float)
target = np.array([0, 1, 1, 0])

feature_means = features.mean(axis=0)
feature_stds = features.std(axis=0)
standardized = (features - feature_means) / feature_stds

print("features shape:", features.shape)
print("target shape:", target.shape)
print("means:", np.round(feature_means, 2))
print("stds:", np.round(feature_stds, 2))
print("standardized features:\n", np.round(standardized, 3))
```

```python
X = np.array([
    [1, 1.0, 2.0],
    [1, 2.0, 1.0],
    [1, 3.0, 4.0]
])
weights = np.array([0.5, 1.2, -0.7])
predictions = X @ weights

print("design matrix X:\n", X)
print("weights:", weights)
print("predictions:", np.round(predictions, 3))
```

## 17. Real Usecase Example

```python
np.random.seed(SEED)
study_hours = np.random.randint(1, 10, size=12)
practice_tests = np.random.randint(50, 100, size=12)
exam_scores = 5 * study_hours + 0.4 * practice_tests + np.random.normal(0, 3, size=12)
student_data = np.column_stack((study_hours, practice_tests, exam_scores))

print("columns = study_hours, practice_tests, exam_scores")
print(np.round(student_data, 2))
print("top 3 exam scores:", np.round(np.sort(student_data[:, 2])[-3:], 2))
```

Best practices:

- Check `shape` before arithmetic and matrix operations.
- Use `.copy()` when you need independence from the original array.
- Be explicit about `dtype` when precision matters.
- Prefer vectorized operations over manual Python loops when possible.
- Print small slices and metadata when debugging instead of staring at full arrays.
